```python
colleges = np.array(diversity['name'].unique())
```

```python
colleges.shape
```

```python
clean_diversity = pd.DataFrame(colleges, columns = ['name'])
```

```python
clean_diversity.head()
```

```python
clean_diversity
```

```python
columns = ['Women', 'American Indian / Alaska Native', 'Asian', 'Black',
       'Hispanic', 'Native Hawaiian / Pacific Islander', 'White',
       'Two Or More Races', 'Unknown', 'Non-Resident Foreign',
       'Total Minority']
```

```python
clean_diversity[columns] = np.nan
```

```python
clean_diversity
```

```python
for index in diversity.index:
    category = diversity.iloc[index]['category']
    name = diversity.iloc[index]['name']
    clean_diversity.loc[clean_diversity['name'] == name, [category]] = diversity.iloc[index]['enrollment']
```

```python
clean_diversity
```

```python
clean_diversity['total_enrollment'] = np.nan
```

```python
clean_diversity
```

```python
for index in diversity.index:
    name = diversity.iloc[index]['name']
    clean_diversity.loc[clean_diversity['name'] == name, ['total_enrollment']] = diversity.iloc[index]['total_enrollment']
```

```python
clean_diversity
```

```python
clean_diversity.to_csv('data/cleaned_diversity_school.csv', index=False)
```
